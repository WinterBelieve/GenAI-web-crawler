```python
import requests
from bs4 import BeautifulSoup

# 目標網站 URL
url = "https://www.cgu.edu.tw/ch/Subject/Detail/96?nodeId=200"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Safari/537.36'
}
response = requests.get(url, headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # 解析分類標題
    categories = {}
    current_category = None
    for header in soup.find_all(['h3']):
        if header.a and header.a.has_attr('title'):
            category_name = header.a['title'].replace('tag to ', '').strip()
            current_category = category_name
            categories[current_category] = []

    # 查找所有包含社團的表格行
    rows = soup.find_all('tr')
    for row in rows:
        # 檢查是否在分類之下
        parent_header = row.find_previous('h3')
        if parent_header and parent_header.a and parent_header.a.has_attr('title'):
            current_category = parent_header.a['title'].replace('tag to ', '').strip()

        # 提取社團名稱 (假設在第一個 <td> 內)
        club_name = row.find('td')
        # 提取社團網址 (假設在第二個 <td> 內的 <a> 標籤內)
        link_tag = row.find('a')

        if club_name and link_tag and current_category:
            club_name_text = club_name.text.strip()
            link = link_tag.get('href')
            categories[current_category].append((club_name_text, link))

    # 列出所有分類和社團
    for category, clubs in categories.items():
        print(f"分類: {category}")
        for club_name, link in clubs:
            print(f"  社團名稱: {club_name}, 網址: {link}")
        print()

else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import csv
import re

# 設定 ChromeDriver 路徑
driver_path = 'D:/chromedriver-win64/chromedriver.exe'  # 替換為你的 chromedriver 路徑
service = Service(driver_path)
browser = webdriver.Chrome(service=service)  # 啟動 Chrome 瀏覽器

try:
    # 打開目標 Dcard 頁面
    browser.get('https://www.dcard.tw/f/cgu/p/240903773')
    
    # 等待頁面加載完成
    time.sleep(5)

    # 抓取目標數據
    elements = browser.find_elements(By.CSS_SELECTOR, "span")

    # 保存數據到 CSV 文件
    with open('club_data.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["社團名稱", "地點"])  # 根據需要調整欄位名稱

        for element in elements:
            text = element.text.strip()
            if "地點" in text:  # 簡單過濾示例，根據實際情況調整
                # 使用正則表達式提取社團名稱和地點
                match = re.search(r'地點｜(.+?)\s*([^\s]+社)', text)
                if match:
                    location = match.group(1).strip()
                    club_name = match.group(2).strip()
                    writer.writerow([club_name, location])
                else:
                    # 如果不符合預期格式，進行進一步處理
                    parts = text.split('地點｜')
                    if len(parts) == 2:
                        location = parts[1].strip()
                        club_name = parts[0].strip() if "社" in parts[0] else "未知社團"
                        writer.writerow([club_name, location])
                    else:
                        print(f"無法處理的文本: {text}")

finally:
    browser.quit()

print("資料抓取完成，請打開 'club_data.csv' 進行人工整理。")
```
